### Eigendecomposition

#### Affine Transformation via Matrix Application

```python
v = np.array([3, 1])
v
```

```python
import matplotlib.pyplot as plt
```

```python
def plot_vectors(vectors, colors):
    """
    Plot one or more vectors in a 2D plane, specifying a color for each.

    Arguments
    ---------
    vectors: list of lists or of arrays
        Coordinates of the vectors to plot. For example, [[1, 3], [2, 2]]
        contains two vectors to plot, [1, 3] and [2, 2].
    colors: list
        Colors of the vectors. For instance: ['red', 'blue'] will display the
        first vector in red and the second in blue.

    Example
    -------
    plot_vectors([[1, 3], [2, 2]], ['red', 'blue'])
    plt.xlim(-1, 4)
    plt.ylim(-1, 4)
    """
    plt.figure()
    plt.axvline(x=0, color='lightgray')
    plt.axhline(y=0, color='lightgray')

    for i in range(len(vectors)):
        x = np.concatenate([[0,0],vectors[i]])
        plt.quiver([x[0]], [x[1]], [x[2]], [x[3]],
                   angles='xy', scale_units='xy', scale=1, color=colors[i],)
```

```python
plot_vectors([v], ['lightblue'])
plt.xlim(-1, 5)
_ = plt.ylim(-1, 5)
```

```python
I = np.array([[1, 0], [0, 1]])
I
```

```python
Iv = np.dot(I, v)
Iv
```

```python
v == Iv
```

```python
plot_vectors([Iv], ['blue'])
plt.xlim(-1, 5)
_ = plt.ylim(-1, 5)
```

#### In contrast, consider this matrix (let's call it E) that flips vectors over the -axis:

```python
E = np.array([[1, 0], [0, -1]])
E
```

```python
Ev = np.dot(E, v)
Ev
```

```python
plot_vectors([v, Ev], ['lightblue', 'blue'])
plt.xlim(-1, 5)
_ = plt.ylim(-3, 3)
```

### Or, this matrix, F, which flips vectors over the -axis:

### In pytorch


```python
A = torch.tensor([[1, 0],[0, 1.]])
A
```

```python
eigens = torch.linalg.eig(A)
eigens
```

```python
lambda_p = eigens.eigenvalues[0]
lambda_p
```

```python
v_p =eigens.eigenvectors[:, 0]
v_p
```

```python
torch.matmul(A, torch.abs(v_p))
```

```python
torch.abs(lambda_p * v_p)
```
